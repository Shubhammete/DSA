```python
class Node:
    def __init__(self,value):
        self.value = value
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None

# Insert Methods

# O(n) linear time complexity to add element to end of list
    def append(self, value):
        if self.head == None:
            self.head = Node(value)
            return
        else:
            last = self.head
            while last.next:
                last = last.next
            last.next = Node(value)

# O(1) time complexity
    def prepend(self,value):
        # firstEle = self.head
        # self.head = Node(value)
        # self.head.next = firstEle
        # # OR
        EleToInsert = Node(value)
        EleToInsert.next = self.head
        self.head = EleToInsert
        

    def insert(self, value, index):
        if self.head == None and index !=0:
            print("List is Empty!")  
            return 
        elif index ==0:
            newNode = Node(value)
            newNode.next = self.head
            self.head = Node(value)
            return
        else:
            last = self.head
            # here we want to reach before the target position and then append it
            for i in range(0,index-1):
                if(last.next):
                    print("Index out of Bound!")
                else:
                    last = last.next
            newNode = Node(value)
            newNode.next = last.next
            last.next = newNode

# --------------------------------------------------------------------------------------------------
# Deletion Methods

    def delete(self, value):
        currentEle = self.head
        while currentEle:
            if(currentEle.next is not None and currentEle.next.value == value):
                currentEle.next = currentEle.next.next
            elif currentEle.next is None:
                currentEle.next = None
                break
            currentEle = currentEle.next

    def pop(self, index):
        currentEle = self.head
        for i in range(0, index - 1):
            currentEle = currentEle.next
        # check for last element
        if currentEle.next:
            currentEle.next = currentEle.next.next
# ------------------------------------------------------------------------------------------
# General Methods

    def contains(self, value):
        if self.head == None:
            print("List is Empty!")

        last = self.head
        while self.head :
            if(last.value == value):
                return True
            last = last.next
        return False
    
    def length(self):
        counter = 0
        last = self.head
        while last.next:
            counter += 1
        return counter
    # instead length we can add counter in append and prepend so each time we add element we increase the counter and vice versa for deletion
    
    def get(self, index):
        currentEle = self.head
        for i in range(0,index):
            currentEle = currentEle.next
        return currentEle.value

    def printList(self):
        currentEle = self.head
        listStr = f"[ {currentEle.value}"
        while currentEle.next:
            # print(currentEle.next.value)
            listStr += f", {currentEle.next.value}"
            currentEle = currentEle.next
        listStr += f" ]"
        return listStr
```
